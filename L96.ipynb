```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class L96(object):
    def __init__(self, K=36, J=10, h=1, F=10, c=10, b=10, dt=0.001):
        self.K, self.J, self.h, self.F, self.c, self.b, self.dt = K, J, h, F, c, b, dt
        self.X = np.random.rand(self.K)
        self.Y = np.zeros((self.K, self.J))
        self.history_X = self.X[None, :]
    def step(self):
        X_rhs = -np.roll(self.X, -1) * (np.roll(self.X, -2) - np.roll(self.X, 1)) - self.X + self.F
        self.X += X_rhs * self.dt
        self.history_X = np.concatenate([self.history_X, self.X[None, :]])
    def iterate(self, steps):
        for n in range(steps):
            self.step()
```

```python
l96 = L96()
```

```python
plt.plot(l96.X)
```

```python
l96.iterate(1000)
```
